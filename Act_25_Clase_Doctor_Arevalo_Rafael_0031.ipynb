```python
# Clase Doctor
class Doctor:
    # Constructor
    def __init__(self, nombre, especialidad, cedula):
        self.nombre = nombre
        self.especialidad = especialidad
        self.cedula = cedula

    # Función para leer datos
    def leerdatos(self):
        self.nombre = input("Nombre del doctor: ")
        self.especialidad = input("Especialidad: ")
        self.cedula = input("Cédula profesional: ")

    # Función para mostrar datos
    def mostrardatos(self):
        print("Nombre:", self.nombre)
        print("Especialidad:", self.especialidad)
        print("Cédula:", self.cedula)


# -------- DOCTOR 1 --------
print("DOCTOR 1")
doctor1 = Doctor("", "", "")
doctor1.leerdatos()

# -------- DOCTOR 2 --------
print("\nDOCTOR 2")
doctor2 = Doctor("", "", "")
doctor2.leerdatos()

# -------- MOSTRAR DATOS --------
print("\nDATOS DOCTOR 1")
doctor1.mostrardatos()

print("\nDATOS DOCTOR 2")
doctor2.mostrardatos()

print("\nPrograma terminado y generado por Arevalo Rafael 0031")
```
